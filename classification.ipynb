```python
#3월에 점수 1을 준다
def month_encoding(month):
    score = ""
    if month.endswith("03") : score = 1
    else : score = 0
    return score
```

```python
book_df["날짜"] = book_df["날짜"].astype(str)
book_df["점수"] = book_df["날짜"].apply(lambda x: month_encoding(x))
book_df
```

```python
import numpy as np

book_df["날짜"] = book_df["날짜"].astype(str)
book_df["소비"] = ""

for i in range(0, 13):
    for j in range(12, 85):
        if (book_df.loc[i,"서적/문구"] >= 0) & (book_df.loc[i,"소매판매지수"] > 0):
            book_df.loc[i,"소비"] = 1
        elif (book_df.loc[i,"서적/문구"] >= 0) & (book_df.loc[i,"소매판매지수"] < 0):
            book_df.loc[i,"소비"] = 1
        elif ((book_df.loc[i,"서적/문구"] < 0) >= book_df["서적/문구"].quantile(q=0.5)) & (book_df.loc[i,"소매판매지수"] > 0):
            book_df.loc[i,"소비"] = 1
        else:
            book_df.loc[i,"소비"] = 0
        if (book_df["서적/문구"].loc[j-12] > 0) & (book_df["서적/문구"].loc[j] > 0):
            book_df.loc[j,"소비"] = 1
        elif (book_df["서적/문구"].loc[j-12] < 0) & (book_df["서적/문구"].loc[j] > 0):
            if ((book_df["서적/문구"].loc[j-12]) >= book_df["서적/문구"].quantile(q=0.5)) & ((book_df.loc[j,"소매판매지수"]) > book_df["소매판매지수"].quantile(q=0.75)):
                book_df.loc[j,"소비"] = 1
            else :
                book_df.loc[j,"소비"] = 0

        elif (book_df["서적/문구"].loc[j-12] > 0) & (book_df["서적/문구"].loc[j] < 0):
            if (book_df["서적/문구"].loc[j] > book_df["서적/문구"].quantile(q = 0.5)) & ((book_df.loc[j,"소매판매지수"]) > book_df["소매판매지수"].quantile(q=0.75)):
                book_df.loc[j,"소비"] = 1
            else:
                book_df.loc[j,"소비"] = 0
        else:
            book_df.loc[j,"소비"] = 0
            

            
print(book_df)
```
