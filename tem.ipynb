```python
import pandas as pd

tem = pd.read_csv('./data/기온.csv', encoding='euc-kr')
tem = tem.drop(['월합강수량(00~24h만)(mm)', '일최다강수량(mm)', '평균풍속(m/s)', '최대풍속(m/s)'], axis = 1)
tem.columns = ['날짜', '평균기온', '평균최고기온', '평균최저기온']
tem['날짜'] = pd.to_datetime(tem['날짜'], format='%Y-%m')
tem['날짜'] = tem['날짜'].dt.strftime('%Y%m')
tem['일교차'] = tem['평균최고기온'] - tem['평균최저기온']
tem['기온등락률'] = tem['평균기온'].pct_change()
tem['일교차등락률'] = tem['일교차'].pct_change()
tem = tem.dropna()
tem
```

```python
tem['날짜'] = tem['날짜'].astype(str)
tem['월'] = tem['날짜'].str[-2:]
tem['월'] = tem['월'].astype(int)
tem.to_csv('./data/tem_df.csv', index= 0)
```

```python
tem_m = tem.groupby('월').mean()
tem_m
```
